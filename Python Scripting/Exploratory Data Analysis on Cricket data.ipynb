```python
import os
import pandas as pd
```

# Combined ODI + Test + T20 stats


```python
base_path = r'D:\Data\Projects\Cricket data\Cricket-Analysis\Dataset\combined odi+ t20 + test stats'
```

```python
# get lit of all .csv files
csv_files = [file for file in os.listdir(base_path) if file.endswith('.csv')]

# Display the list
csv_files
```

```python
# Dictionary to hold Datframe
df_combined = {}

# Read each csv file into dataframe
for file in csv_files:
    file_path = os.path.join(base_path, file)
    # Remove 'combined_' prefix and '.csv' suffux for cleaner key names
    key = file.replace('.csv', '')
    df_combined[key]= pd.read_csv(file_path)

# show key loaded
list(df_combined.keys())
```

```python
for key in df_combined:
    print(f"\nðŸ“Š Data for: {key}")
    display(df_combined[key])
```

```python
# removing NaN frpm tables
for key in df_combined:
    df_combined[key].dropna(how="any", inplace=True)
```

```python
# Get the specific DataFrame, as we are going to add full name as a column
df_Team = df_combined['combined_results_summary']
df_Team
```

```python
# Get the specific DataFrame, as we are going to add full name as a column
df_Team = df_combined['combined_results_summary']

# Define the team code to full name mapping
team_name_map = {
    'Eng': 'England',
    'Aus': 'Australia',
    'Ind': 'India',
    'Pak': 'Pakistan',
    'WI': 'West Indies',
    'NZ': 'New Zealand',
    'SL': 'Sri Lanka',
    'SA': 'South Africa',
    'Zim': 'Zimbabwe',
    'Ban': 'Bangladesh',
    'Ire': 'Ireland',
    'Afgh': 'Afghanistan',
    'Scot': 'Scotland',
    'Ken': 'Kenya',
    'Neth': 'Netherlands'
}
# Add a new column for the full team name
df_Team['Team_full'] = df_Team['Team'].map(team_name_map)

#save back to the main directory
df_combined['combined_results_summary'] = df_Team

#previe updayed data frame
display(df_combined)
```

```python

# Define the output Excel file path
output_excel_path = "D:\Data\Projects\Cricket data\Cricket-Analysis\Dataset\combined odi+ t20 + test stats\combined_odi_t20_test_stats.xlsx"

# Create a Pandas Excel writer using openpyxl engine
with pd.ExcelWriter(output_excel_path, engine='openpyxl') as writer:
    for sheet_name, df in df_combined.items():
        # Clean sheet name (Excel doesn't allow some characters)
        clean_sheet_name = sheet_name[:31].replace('/', '_')  # Excel sheet name max = 31 chars
        df.to_excel(writer, sheet_name=clean_sheet_name, index=False)

print(f"âœ… Excel file created: {output_excel_path}")
```
